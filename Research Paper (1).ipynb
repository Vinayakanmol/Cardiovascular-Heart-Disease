# Data Pre-processing

```python
#Splitting the data 
X=df.drop(['DEATH_EVENT'],axis=1)
y=df['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Data Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
accuracy_list=[]
```

# Model Building

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score, roc_curve
lr=LogisticRegression()
prediction=lr.fit(X_train,y_train)
y_prediction=prediction.predict(X_test)
accuracy=(round(accuracy_score(y_prediction,y_test),4)*100)
accuracy_list.append(accuracy)
```

```python
print('Classification report for Logistic Regression model is:')
cr=classification_report(y_test,y_prediction)
print(cr)
print(' ')
roc=roc_auc_score(y_test,y_prediction)
print('The ROC score for Logistic Regression is:',roc)
print(' ')
cm=confusion_matrix(y_test,y_prediction)
print('The confusion matrix for Logistic Regression is:')
print(cm)
print(' ')
score=accuracy_score(y_test,y_prediction)
print('The accuracy we got for Logistic Regression is:',score)
```

```python
sns.heatmap(cm,annot=True,fmt='d',cmap='Greens',
           xticklabels=['Predictive Negative','Predictive Positive'],
           yticklabels=['Actual Negative','Actual Postivie'])
plt.title('(Figure 13) Confusion Matrix for Logistic Regression')
```

# Support Vector Machine 

```python
from sklearn.svm import SVC
svmm=SVC()
prediction1=svmm.fit(X_train,y_train)
y_prediction1=prediction1.predict(X_test)
accuracy1=(round(accuracy_score(y_test,y_prediction1),4)*100)
accuracy_list.append(accuracy1)
```

```python
print('Classification report for SVM model is:')
cr1=classification_report(y_test,y_prediction1)
print(cr1)
print(' ')
roc1=roc_auc_score(y_test,y_prediction1)
print('The ROOC score of SVM model is:',roc1)
print(' ')
cm1=confusion_matrix(y_test,y_prediction1)
print('The confusion matrix of SVM is:')
print(cm1)
print(' ')
print('The accuracy score of SVM model is:',accuracy_score(y_test,y_prediction1))
```

# K-Nearest Neigbhors

```python
#Finding the best k value
from sklearn.neighbors import KNeighborsClassifier
knn_list=[]
for k in range(1,50):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    knn_pred=knn.predict(X_test)
    knn_accuracy=(round(accuracy_score(y_test,knn_pred),4)*100)
    knn_list.append(knn_accuracy)
k=np.arange(1,50)
plt.plot(k,knn_list)
```
